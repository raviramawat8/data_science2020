```python
import requests

url = "https://covid-19-india-data-by-zt.p.rapidapi.com/GetIndiaHistoricalDataBetweenDates"

querystring = {"start_date":"2020-01-01","end_date":"2020-05-01"}

headers = {
    'x-rapidapi-host': "covid-19-india-data-by-zt.p.rapidapi.com",
    'x-rapidapi-key': "7cd6bf113fmsh7358008b99e2b68p131b68jsnf29f543d01fd"
    }

response = requests.request("GET", url, headers=headers, params=querystring)

print(response.text)
```

```python
d=response.json()
```

```python
len(d['records'])
```

```python
for var in range(len(d['records'])):
    print(d['records'][var])
```

```python
for var in range(len(d['records'])):
    for key,value in d['records'][var].items():
        print(key,'=',value)
```

```python
date=[]
for var in range(len(d['records'])):
    for key,value in d['records'][var].items():
        if key=='dateofrecord':
            date.append(value)
```

```python
totaldeath=[]
for var in range(len(d['records'])):
    for key,value in d['records'][var].items():
        if key=='cases':
            totaldeath.append(int(value['totaldeceased']))
```

```python
totalconfirmed=[]
for var in range(len(d['records'])):
    for key,value in d['records'][var].items():
        if key=='cases':
            totalconfirmed.append(int(value['totalconfirmed']))
```

```python
totalrecover=[]
for var in range(len(d['records'])):
    for key,value in d['records'][var].items():
        if key=='cases':
            totalrecover.append(int(value['totalrecovered']))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.DataFrame(data=[totalconfirmed,totalrecover,totaldeath])
```

```python
df=df.T
```

```python
df.columns=['totalconfirmed','totalrecovered','totaldeath']
```

```python
df.index=[date]
```

```python
df
```

```python
fig = plt.figure(figsize=(15,4))
ax = fig.add_subplot(111)
ax.scatter(date[::3],df['totalconfirmed'][::3],c='r',label="Total Confirmed")
ax.plot(date[::3],df['totalconfirmed'][::3],'k--')
ax.scatter(date[::3],df['totalrecovered'][::3],c='g',alpha=0.7,label="Total Recovered")
ax.plot(date[::3],df['totalrecovered'][::3],'k--',alpha=0.7)
ax.scatter(date[::3],df['totaldeath'][::3],c='k',alpha=0.5,label="Total Death")
ax.plot(date[::3],df['totaldeath'][::3],'r--',alpha=0.5)
ax.set_xticklabels(date[::3],rotation=80)
ax.grid(color='blue',ls='-.',alpha=0.4)
ax.set_title("COVID-19 INDIA",fontsize=20)
ax.set_xlabel('Dates')
ax.set_ylabel('Counts')
plt.legend()
plt.show()
```

```python
data['data']=data['data'][3:]
```

```python
for var in range(len(data['data'])):
    print(data['data'][var])
```

```python
district = []
for var in range(len(data['data'])):
    district.append(data['data'][var]['name'])
```

```python
active = []
for var in range(len(data['data'])):
    active.append(int(data['data'][var]['active']))
```

```python
death = []
for var in range(len(data['data'])):
    death.append(int(data['data'][var]['deceased']))
```

```python
confirmed = []
for var in range(len(data['data'])):
    confirmed.append(int(data['data'][var]['confirmed']))
```

```python
recovered = []
for var in range(len(data['data'])):
    recovered.append(int(data['data'][var]['recovered']))
```

```python
df=pd.DataFrame(data=[district,confirmed,active,recovered,death])
```

```python
df=df.T
```

```python
df.columns=['district','confirmed','active','recovered','death']
```

```python
df
```

```python
fig = plt.figure(figsize=(14,4))
plt.plot('district','confirmed','ro--',data=df)
plt.plot('district','recovered','go--',data=df)
plt.plot('district','death','ko--',data=df)
plt.xticks(rotation=80)
plt.yticks(np.arange(0,1001,50))
plt.grid(c='b',ls='--',alpha=0.5)
plt.show()
```

```python
fig = plt.figure(figsize=(16,5))
a1 = fig.add_subplot(111)
x = np.arange(len(df['district']))
a1.bar(x+0.0,height=df['confirmed'],ec='k',width=0.2,label='Confirmed')
a1.bar(x+0.2,height=df['active'],width=0.2,ec='k',label='Active')
a1.bar(x+0.4,height=df['recovered'],width=0.2,ec='k',label='Recovered')
a1.bar(x+0.6,height=df['death'],width=0.2,ec='k',label='Death')
plt.legend()
a1.set_xticks(np.arange(len(df['district'])))
a1.set_xticklabels(df['district'],rotation=80,fontsize=13)
a1.set_yticks(np.arange(0,1001,50))
plt.grid(c='b',ls='--',alpha=0.5)
plt.show()
```
