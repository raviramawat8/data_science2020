```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig =plt.figure(figsize=(10,5))
a1 = fig.add_subplot(111)
a2 = fig.add_subplot(4,4,16)
a = np.linspace(0,np.pi*2,360)
b = np.array(list(a[:70])+list(a[110:240])+list(a[290:]))
a1.plot(np.sin(a),'b-.')
a1.set_title("Trigonometric Parameters",fontsize=20)
a1.set_xlabel("X-axis")
a1.set_ylabel("Y-axis")
a1.set_xlim(-50,410)
a1.set_ylim(-5,5)
a1.grid()
a1.set_xticks([0,90,180,270,360])
a2.plot(np.sin(a),'b-.')
a2.grid()
a1.plot(np.cos(a),'g-.')
a1.plot(np.tan(b),'r--')
a1.plot([0,360],[0,0],'k')
plt.show()
```

```python
from random import randint
```

```python
fig = plt.figure(figsize=(14,5))
axes = plt.axes()
axes.set_xlim(0,100)
axes.set_ylim(0,100)
axes.set_xticks([var for var in range(-10,111,5)])
axes.set_yticks([var for var in range(-10,111,15)])
fig.set_facecolor('gray')
x = np.array([var for var in range(0,100,2)])
y = np.array([randint(1,100) for var in range(0,100,2)])
z = np.array([randint(1,100) for var in range(0,100,2)])
plt.scatter(x,y,c='b')
plt.scatter(x,z,c='g')
plt.plot(x,y,'b:',linewidth=2)
plt.plot(x,z,'g--',linewidth=2)
plt.xlabel('Days')
plt.ylabel('Students')
plt.legend(['Growth2018','Growth2019'],loc=1)
plt.title("Grras Solution Pvt. Ltd.",fontsize=20)
plt.grid(color='r',linestyle='--')
plt.show()
```

```python
#stack plot
days= [1,2,3,4,5]
working=[7,8,7,2,3]
sleeping=[5,7,3,2,1]
eating = [1,2,3,1.2,0.5]
playing=[6,4,20,10,13]

plt.plot([],[],color='r',label='days',linewidth=2)
plt.plot([],[],color='b',label='working',linewidth=2)
plt.plot([],[],color='g',label='sleeping',linewidth=2)
plt.plot([],[],color='k',label='eating',linewidth=2)
plt.plot([],[],color='m',label='playing',linewidth=2)
plt.stackplot(days,sleeping,working,eating,playing,colors=['r','g','b','k','m'])
plt.legend()
plt.show()
```

```python
import time
import psutil
%matplotlib notebook
plt.rcParams['animation.html'] = 'jshtml'
fig=plt.figure(figsize=(10,4))
ax = fig.add_subplot(111)
fig.show()
```

```python
i =0
x ,y = [],[]
while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    ax.plot(x,y,color='b')
    fig.canvas.draw()
    ax.set_xlim(left=max(0,i-50),right=i+50)
    time.sleep(0.1)
    i+=1
```
