```python
df = pd.read_csv('c:/Users/HP-PC/Downloads/sk.csv')
```

```python
df.plot(kind='bar')
```

```python
fig = plt.figure(figsize=(14,4))
plt.bar(df['STATE NAME'],df['TOTAL CONFIRMED'],color='maroon',ec='k',label="Total Confirmed")
plt.bar(df['STATE NAME'],df['CURED/DISCHARGED/MIGRATED'],color='Green',ec='k',label="Recovered")
plt.bar(df['STATE NAME'],df['DEATH'],color='k',ec='k',label="Death")
plt.xticks(rotation=80)
plt.legend()
plt.show()
```

```python
df.sort_values(by="TOTAL CONFIRMED",inplace=True,ascending=False)
```

```python
fig = plt.figure(figsize=(20,7))
x= np.arange(31)
ax=fig.add_axes([0,0,1,1])
plt.bar(x+0.0,df['TOTAL CONFIRMED'],color='maroon',ec='k',label="Total Confirmed",width=0.5)
plt.bar(x+0.2,df['CURED/DISCHARGED/MIGRATED'],color='Green',ec='k',label="Recovered",width=0.5)
plt.bar(x+0.4,df['DEATH'],color='k',ec='k',label="Death",width=0.5)
plt.xticks(x,rotation=80)
ax.set_xticklabels(df['STATE NAME'])
plt.legend()
plt.show()
```
